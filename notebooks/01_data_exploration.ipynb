# **Thông tin về bộ dữ liệu**

- **Đường dẫn đến dữ liệu:** [Credit card customer](https://www.kaggle.com/datasets/sakshigoyal7/credit-card-customers)

- **Bộ dữ liệu bao gồm 23 cột với thông tin chi tiết như sau:**
    <ol>
        <li> <b>CLIENTNUM</b>: Cột gồm mã số định danh duy nhất dành cho mỗi khách hàng.
        <li> <b>Attrition_Flag</b>: Phản ánh trạng thái hoạt động của khách hàng. Cột này bao gồm hai giá trị phân loại chính là 'Attrited Customer' (Khách hàng rời bỏ) và 'Existing Customer' (Khách hàng hiện hữu).
        <li> <b>Customer_Age</b>: Biểu thị độ tuổi của khách hàng (tính bằng năm). Dữ liệu ghi nhận các khách hàng là những người trưởng thành, với độ tuổi dao động từ mức thấp nhất là 26 đến cao nhất là 73 tuổi.
        <li> <b>Gender</b>: Cột chỉ giới tính của khách hàng. Dữ liệu bao gồm hai giá trị phân loại là 'M' (Nam) và 'F' (Nữ).
        <li> <b>Dependent_count</b>: Cột cho biết số lượng người phụ thuộc vào chủ tài khoản. Số lượng này là các số nguyên nằm trong khoảng từ 0 đến 5 người.
        <li> <b>Education_Level</b>: Mô tả trình độ học vấn của chủ tài khoản. Các mức độ học vấn được ghi nhận bao gồm: 'College', 'Doctorate', 'Graduate', 'High School', 'Post-Graduate', 'Uneducated' và 'Unknown'.
        <li> <b>Marital_Status</b>: Thể hiện tình trạng hôn nhân của khách hàng. Các trạng thái bao gồm: 'Divorced' (Ly hôn), 'Married' (Đã kết hôn), 'Single' (Độc thân) và 'Unknown' (Không xác định).
        <li> <b>Income_Category</b>: Phân loại mức thu nhập hàng năm của chủ tài khoản. Thu nhập được chia thành các nhóm: 'Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +' và 'Unknown'.
        <li> <b>Card_Category</b>: Biến sản phẩm cho biết loại thẻ tín dụng mà khách hàng đang sử dụng. Các hạng thẻ bao gồm: 'Blue', 'Gold', 'Platinum' và 'Silver'.
        <li> <b>Months_on_book</b>: Thời gian quan hệ giữa khách hàng với ngân hàng, tính bằng tháng. Dữ liệu cho thấy thời gian gắn bó của khách hàng kéo dài từ 13 tháng đến 56 tháng.
        <li> <b>Total_Relationship_Count</b>: Tổng số lượng sản phẩm hoặc dịch vụ mà khách hàng đang nắm giữ tại ngân hàng. Số lượng sản phẩm này dao động từ 1 đến 6 sản phẩm.
        <li> <b>Months_Inactive_12_mon</b>: Số tháng mà tài khoản không có hoạt động nào trong vòng 12 tháng qua. Số liệu này nằm trong khoảng từ 0 đến 6 tháng.
        <li> <b>Contacts_Count_12_mon</b>: Số lần khách hàng liên hệ với ngân hàng trong 12 tháng gần nhất. Tần suất liên hệ được ghi nhận từ 0 đến 6 lần.
        <li> <b>Credit_Limit</b>: Hạn mức tín dụng tối đa trên thẻ của khách hàng. Giá trị hạn mức là các số thực, trải rộng từ 1,438.3 đến 34,516.0 đơn vị tiền tệ.
        <li> <b>Total_Revolving_Bal</b>: Tổng số dư nợ quay vòng (số tiền nợ chưa thanh toán hết và chuyển sang kỳ sau) trên thẻ tín dụng. Số dư này dao động từ 0 đến 2,517.0.
        <li> <b>Avg_Open_To_Buy</b>: Hạn mức tín dụng khả dụng trung bình để mua sắm (tính trung bình trong 12 tháng qua). Số tiền khả dụng này nằm trong khoảng từ 3.0 đến 34,516.0.
        <li> <b>Total_Amt_Chng_Q4_Q1</b>: Tỷ lệ thay đổi trong tổng số tiền giao dịch khi so sánh Quý 4 với Quý 1. Tỷ lệ này là các giá trị số thực nằm trong khoảng từ 0.0 đến 3.397.
        <li> <b>Total_Trans_Amt</b>: Tổng số tiền giao dịch mà khách hàng đã thực hiện trong 12 tháng qua. Giá trị giao dịch nằm trong khoảng từ 510.0 đến 18,484.0.
        <li> <b>Total_Trans_Ct</b>: Tổng số lần (lượt) giao dịch khách hàng thực hiện trong 12 tháng qua. Số lượng giao dịch dao động từ 10 đến 139 lần.
        <li> <b>Total_Ct_Chng_Q4_Q1</b>: Tỷ lệ thay đổi trong tổng số lượng giao dịch khi so sánh Quý 4 với Quý 1. Tỷ lệ này bao gồm các giá trị từ 0.0 đến 3.714.
        <li> <b>Avg_Utilization_Ratio</b>: Tỷ lệ sử dụng thẻ trung bình của khách hàng (tỷ lệ giữa số dư nợ và hạn mức tín dụng). Tỷ lệ này dao động từ 0.0 đến 0.999.
        <li> <b>Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1</b>: Đây là cột chứa kết quả từ thuật toán Naive Bayes được sử dụng trong quá trình xây dựng dữ liệu gốc.
        <li> <b>Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2</b>: Tương tự như cột trên, đây là một biến phụ sinh ra từ thuật toán Naive Bayes.

# Library used to exploring data

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_path = '../data/BankChurners.csv'
data = np.genfromtxt(csv_path, delimiter=',', dtype=str, encoding='ascii')

# Loại bỏ hai cột cuối không cần thiết
data = data[:, :-2]
columns = data[0, :]
columns = np.char.replace(columns, '"', '')
data = data[1:, :]
data = np.char.replace(data, '"', '')
```

# I. Phân tích một số thông tin cơ bản của bộ dữ liêu

## I.1 Số chiều của bộ dữ liệu
- Xác định số dòng và số cột của bộ dữ liệu.

```python

print("Số dòng:", data.shape[0])
print("-" * 40)
print("Số cột:", data.shape[1])
print("-" * 40)
# Tên các cột
print("Tên các cột:")
print("-" * 40)
for i, col in enumerate(columns):
    print(f"\tCột {i}: {col}")

categorical_columns = ['CLIENTNUM', 'Attrition_Flag', 'Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category']
numerical_columns = [col for col in columns if col not in categorical_columns]
print("Chúng ta có", len(categorical_columns), "cột phân loại và", len(numerical_columns), "cột số.")
print("-" * 40)
print("\tCác cột phân loại:", categorical_columns)
print("\tCác cột số:", numerical_columns)
```

## I.2 Khoảng giá trị của từng cột
- Xác định số lượng các giá trị khác nhau của từng cột: đối với cột gồm các giá trị có mục đích phân loại thì in ra các `các giá trị riêng biệt`, còn đối với các cột có giá trị gồm nhiều số thì in ra `số lớn nhất, số nhỏ nhất, số trung bình, trung vị và phương sai`.


```python
print("Thống kê mô tả cho từng cột:")
for i, col in enumerate(columns):
    col_data = data[:, i]
    if col in categorical_columns:
        if (col == 'CLIENTNUM'):
            print(f"Cột '{col}' (Phân loại) - Bỏ qua giá trị duy nhất cho CLIENTNUM (cột định danh).")
            continue
        unique_values = np.unique(col_data)
        print(f"Cột '{col}' (Phân loại) - Giá trị duy nhất: {unique_values}")
    else:
        col_data = col_data.astype(float)
        print(f"Cột '{col}' (Số) - Số nhỏ nhất: {np.min(col_data)}, Số lớn nhất: {np.max(col_data)}, Trung bình: {np.mean(col_data):.2f}, Trung vị: {np.median(col_data):.2f}, Độ lệch chuẩn: {np.std(col_data):.2f}")
```

# II. Trực quan hóa dữ liệu

### Xây dựng dictionary để mapping nhanh dữ liệu

```python
dict_data = {'CLIENTNUM': 0, 
             'Attrition_Flag': 1,
             'Customer_Age': 2,
             'Gender': 3,
             'Dependent_count': 4, 
             'Education_Level': 5, 
             'Marital_Status': 6,
             'Income_Category': 7, 
             'Card_Category': 8, 
             'Months_on_book': 9,
             'Total_Relationship_Count': 10, 
             'Months_Inactive_12_mon': 11, 
             'Contacts_Count_12_mon': 12,
             'Credit_Limit': 13, 
             'Total_Revolving_Bal': 14, 
             'Avg_Open_To_Buy': 15, 
             'Total_Amt_Chng_Q4_Q1': 16,
             'Total_Trans_Amt': 17, 
             'Total_Trans_Ct': 18, 
             'Total_Ct_Chng_Q4_Q1': 19, 
             'Avg_Utilization_Ratio': 20}
```

## II.1 Phân phối giá trị của từng cột dữ liệu

Đối với các cột số thì dùng `biểu đồ tần số (Histogram)`, còn đối với các cột phân loại thì dùng `biểu đồ cột (Bar chart)`

```python
cols_to_plot = columns[columns != "CLIENTNUM"]
num_plots = len(cols_to_plot)
ncols = 3
nrows = (num_plots + ncols - 1) // ncols
fig, axes = plt.subplots(nrows, ncols, figsize=(18, nrows * 4))
axes = axes.flatten()

for idx, col in enumerate(cols_to_plot):
    ax = axes[idx]
    col_values = data[:, dict_data[col]]

    if col in categorical_columns:

        unique_vals, counts = np.unique(col_values.astype(str), return_counts=True)

        ax.bar(
            np.arange(len(unique_vals)),
            counts,
            color=sns.color_palette("Set2", len(unique_vals)),
            edgecolor="black"
        )

        ax.set_xticks(np.arange(len(unique_vals)))
        ax.set_xticklabels(unique_vals, rotation=45, ha="right")
        ax.set_ylabel("Count")
        ax.set_title(f"{col} (Categorical)")

    else:
        col_raw = col_values.astype(object)
        mask = col_raw != ""
        numeric = np.full(col_raw.shape, np.nan)
        numeric[mask] = col_raw[mask].astype(float)

        ax.hist(
            numeric,
            bins=30,
            density=True,
            alpha=1.0,
            color="lightblue",
            edgecolor="black",
            linewidth=0.5
        )

        valid_numeric = numeric[~np.isnan(numeric)]
        if valid_numeric.size > 1:
            sns.kdeplot(
                valid_numeric,
                ax=ax,
                color="red",
                linewidth=2,
                clip=(valid_numeric.min(), valid_numeric.max()),
                bw_method="scott"
            )

        ax.set_title(f"{col} (Numerical)")
        ax.set_xlabel(col)
        ax.set_ylabel("Density")


for ax in axes[num_plots:]:
    ax.axis("off")

plt.tight_layout()
plt.subplots_adjust(hspace=0.6, bottom=0.06)
plt.show()
```

## Phân tích dữ liệu có được từ Histogram & Bar Chart

Dựa trên các biểu đồ trực quan hóa, dưới đây là nhận xét chi tiết về đặc điểm phân phối của từng cột dữ liệu:
- **Attrition_Flag (Cột mục tiêu):** Số lượng "Existing Customer" (Khách hàng hiện tại) áp đảo hoàn toàn so với "Attrited Customer" (Khách hàng rời bỏ).

- **Gender (Giới tính):** Nữ giới (F) có số lượng nhỉnh hơn một chút so với Nam giới (M) nhưng xét về tổng thể thì hai bên tương đối bằng nhau.

- **Education_Level (Trình độ học vấn):** Từ đồ thị ta có thể thấy mhóm "Graduate" (Tốt nghiệp ĐH) chiếm tỷ trọng cao nhất và nhóm "Doctorate" chiếm ỷ trọng thấp nhất. Trong đó, có một lượng đáng kể dữ liệu là "Unknown" (Chưa xác định).

- **Marital_Status (Tình trạng hôn nhân):** Qua đồ thị, nhóm "Married" chiếm tỷ trọng cao nhất, tiếp theo là "Single" và cuối cùng, "Divorced" và "Unknown" là các nhóm chiếm tỷ trọng thất nhất.

- **Income_Category (Nhóm thu nhập):** Hầu hết người dùng đều phân bố vào các nhóm có thu nhập dưới `$120k`, đặc biệt là nhóm `dưới $40K` chiếm tỷ trọng khá lớn.

- **Card_Category (Loại thẻ):** Loại thẻ "Blue" chiếm đa số tuyệt đối (có thể >90%). Các loại thẻ cao cấp (Gold, Silver, Platinum) chiếm tỷ lệ không đáng kể. 

- **Customer_Age (Tuổi khách hàng):** Dữ liệu có hình chuông đối xứng đẹp nên ta có thể kết luận phân phối trong cột này là phân phối chuẩn. Độ tuổi trung bình tập trung khoảng 40-50 tuổi.

- **Dependent_count (Số người phụ thuộc):** Phân phối hơi lệch phải nhẹ, gần với phân phối chuẩn. Đỉnh (Mode) nằm ở mức 2 và 3 người phụ thuộc.

* **Months_on_book (Thời gian quan hệ tín dụng):** Phân phối chuẩn với độ nhọn cao (High Kurtosis).Có một đỉnh cực cao ở giá trị **36 tháng**. Điều này cho thấy đa số khách hàng đã sử dụng dịch vụ được tròn 3 năm.

- **Total_Relationship_Count (Tổng số sản phẩm sử dụng)** Phân phối khá đồng đều ở mức cao. Khách hàng thường sử dụng từ 3 đến 6 sản phẩm. Số lượng người dùng 1-2 sản phẩm thấp hơn.

- **Months_Inactive_12_mon (Số tháng không hoạt động trong 1 năm):** Tập trung ở mức thấp. Phổ biến nhất là 1, 2, hoặc 3 tháng không hoạt động. Rất hiếm khách hàng không hoạt động trên 4 tháng.

- **Contacts_Count_12_mon (Số lần liên hệ trong 1 năm):** Phân phối gần chuẩn. Đỉnh nằm ở 2 và 3 lần liên hệ.

- **Credit_Limit (Hạn mức tín dụng)**: Phân phối Lệch phải (Right-skewed). Đa số khách hàng có hạn mức thấp (< 5.000). Có một cái đuôi dài về phía giá trị cao. Đặc biệt có một đỉnh nhỏ bất thường ở giá trị tối đa (khoảng 33k-34k).

- **Total_Revolving_Bal (Tổng dư nợ xoay vòng):** Phân phối hai đỉnh (Bimodal).Có một lượng lớn khách hàng có dư nợ bằng 0 (thanh toán hết sao kê). Đỉnh thứ hai tập trung quanh mức 1.500 - 2.500.

- **Avg_Open_To_Buy (Hạn mức khả dụng trung bình):** Phân phối lệch phải (tương tự Credit Limit). Vì `Avg_Open_To_Buy` = `Credit_Limit` - `Balance`, và nhiều người có Balance = 0, nên biểu đồ này rất giống biểu đồ Credit Limit.

- **Total_Amt_Chng_Q4_Q1 (Tỷ lệ thay đổi số tiền giao dịch Q4 so với Q1):** Phân phối chuẩn, tập trung quanh mức dưới 1.0. Đa số khách hàng có số tiền giao dịch Q4 thấp hơn Q1 một chút.

- **Total_Trans_Amt (Tổng số tiền giao dịch):** Phân phối đa đỉnh (Multimodal) & Lệch phải. Có các nhóm khách hàng chi tiêu khác nhau: nhóm thấp (~2.000), nhóm trung bình (~4.500) và một nhóm nhỏ chi tiêu rất cao (~15.000).

- **Total_Trans_Ct (Tổng số lần giao dịch):** Phân phối hai đỉnh. Một nhóm giao dịch khoảng 40 lần/năm, nhóm tích cực hơn giao dịch khoảng 70-80 lần/năm.

- **Total_Ct_Chng_Q4_Q1 (Tỷ lệ thay đổi số lần giao dịch Q4 so với Q1):** Phân phối chuẩn, hơi lệch. Tương tự như thay đổi số tiền, tỷ lệ này tập trung quanh 0.6 - 0.8.

- **Avg_Utilization_Ratio (Tỷ lệ sử dụng hạn mức):** Phân phối lệch phải mạnh. Đỉnh cao nhất tại 0 (tương ứng với nhóm không nợ tiền). Phần còn lại trải dài với mật độ thấp dần về phía 1.0 (full hạn mức).

## II.2 Phần trăm các giá trị trong các cột phân loại 

Sử dụng `biểu đồ tròn (Pie chart)` để hiển thị phần trăm các giá trị trong các cột phân loại

### Lấy các cột phân loại để vẽ biểu đồ tròn

```python
categorical_columns = np.array(categorical_columns)
categorical_columns_pie = categorical_columns[categorical_columns != 'CLIENTNUM']
print(categorical_columns_pie)
```

```python
# Vẽ biểu đồ tròn cho các cột phân loại
categorical_columns = np.array(categorical_columns)
categorical_columns_pie = categorical_columns[categorical_columns != 'CLIENTNUM']

num_cols = len(categorical_columns_pie)
num_rows = (num_cols + 2) // 3


fig, axes = plt.subplots(num_rows, 3, figsize=(18, num_rows * 5))
axes = axes.flatten()

for i, col in enumerate(categorical_columns_pie):
    # Xử lý chuỗi bằng numpy (vector hóa)
    col_data = np.char.strip(data[:, dict_data[col]], '"')
    # Lấy unique và đếm nhanh
    unique_vals, counts = np.unique(col_data, return_counts=True)

    axes[i].pie(counts, labels=unique_vals, autopct='%1.1f%%', startangle=140)
    axes[i].set_title(f'Phân phối {col} (Phân loại)')

plt.tight_layout()
plt.show()
```

## Phân tích dữ liệu từ biểu đồ tròn (Pie Chart)

    - **Months_Inactive_12_mon (+0.15):** Thời gian "ngủ đông" càng dài, nguy cơ rời bỏ càng lớn. Khách hàng không phát sinh giao dịch trong nhiều tháng liên tiếp là nhóm rủi ro cao.

- **Nhóm biến ít tương quan (Neutral):**
*Hệ số tương quan xấp xỉ 0 (Màu nhạt/Trắng).*
    - Các biến này bao gồm: **Customer_Age**, **Dependent_count**, **Months_on_book**, **Credit_Limit** và **Avg_Open_To_Buy**.
    - Hệ số tương quan rất thấp (gần bằng 0) cho thấy các yếu tố về nhân khẩu học (tuổi, người phụ thuộc), thâm niên hay năng lực tài chính (hạn mức khả dụng) **gần như không có mối quan hệ tuyến tính trực tiếp** đến việc khách hàng đi hay ở. Việc có nhiều tiền (hạn mức cao) hay ít tiền không quyết định việc họ trung thành với ngân hàng.

### II.3. 1 Ma trận tương quan giữa các cột có giá trị phân loại với cột đăc trưng Attrition_Flag


```python
# Compute Cramér's V using fully vectorized contingency table
def cramers_v_fast(a, b):
    # Encode categories as integers
    a_unique, a_enc = np.unique(a, return_inverse=True)
    b_unique, b_enc = np.unique(b, return_inverse=True)

    # Build contingency table using bincount (no loop)
    contingency = np.bincount(
        a_enc * len(b_unique) + b_enc,
        minlength=len(a_unique) * len(b_unique)
    ).reshape(len(a_unique), len(b_unique))

    # χ² calculation (vectorized)
    n = contingency.sum()
    row_sum = contingency.sum(axis=1, keepdims=True)
    col_sum = contingency.sum(axis=0, keepdims=True)
    expected = row_sum @ col_sum / n

    chi2 = ((contingency - expected)**2 / expected).sum()
    phi2 = chi2 / n

    r, k = contingency.shape
    return np.sqrt(phi2 / min(r - 1, k - 1))

categorical_columns = np.array(categorical_columns)
categorical_columns = categorical_columns[categorical_columns != 'CLIENTNUM']

n = len(categorical_columns)
cramers_v_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        if i == j:
            cramers_v_matrix[i, j] = 1
        else:
            cramers_v_matrix[i, j] = cramers_v_fast(data[:, dict_data[categorical_columns[i]]], data[:, dict_data[categorical_columns[j]]])

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    cramers_v_matrix,
    annot=True, fmt=".2f",
    xticklabels=categorical_columns,
    yticklabels=categorical_columns,
    cmap="coolwarm", square=True
)
plt.title("Cramér's V Heatmap Between Categorical Columns")
plt.tight_layout()
plt.show()
```
